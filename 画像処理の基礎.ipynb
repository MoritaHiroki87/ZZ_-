# 8.3 グレースケール変換

RGBの三色で表していたのが白黒の一色（一つのグラデーション）で表現できるようになったので、画像のshapeが変わる。

```
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```
plt.imshow(img_gray)
```

```
plt.gray()
plt.imshow(img_gray)
```

```
img_gray.shape
```
